## Importing pandas library

```python
import pandas as pd
```

### Reading the dataset and creating pandas DataFrame

```python
df_ASN = pd.read_csv('dataset.csv')
```

```python
df_ASN.head()
```

## Throughput = Total data sent / Total time taken to send the data

```python
df_ASN['Throughput(KB/sec)'] = ((df_ASN['Bytes']/df_ASN['Send/Time'])/1024).round(4)
```

```python
df_Co_Server = df_ASN.copy()
ASN_group = df_ASN.groupby('ASN')
Co_Server_group = df_Co_Server.groupby('Co Server')
```

## Average throughput for ASN

```python
df_ASN['Average Throughput(KB/sec)'] = ASN_group['Throughput(KB/sec)'].transform('mean')
df_ASN
```

## Average throughput for Co Server

```python
df_Co_Server['Average Throughput(KB/sec)'] = Co_Server_group['Throughput(KB/sec)'].transform('mean')
df_Co_Server
```
